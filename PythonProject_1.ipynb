```python
import pandas as pd
import sqlite3
import os

# create the sqlite DB
db = sqlite3.connect("MyDemo.db")
```

```python
# ---- Also note, changed data set from URL to relative path -----

# read CSV into dataframe
df = pd.read_csv(r"demo_AmexTransactions.csv", header = None)
```

```python
# convert the pandas dataframe to a sqlite table
df.to_sql("expenses", db, if_exists="replace")
```

```python
# function for SQL query
def run_query(query):
    return pd.read_sql_query(query, db)
```
